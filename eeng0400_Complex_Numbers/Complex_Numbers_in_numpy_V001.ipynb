## Create a range of angles $\phi$ with fractions of $\pi$

```python
N = 3 # number or periods
rv = np.arange(-N,N+1e-9, 0.02) # arange, from -N to +N in steps 0.1
phi = rv * pi # arange, from -N*pi tp +N*pi 
#print(f"rv:\n {rv}")
#print(f"phi:\n {phi}")
```

```python
A = 1.
s = A*sin(phi)
s_p = A*sin(phi+0.5*pi)
c = A*cos(phi)
```

```python
fig1, ax1 = plt.subplots(figsize=(6,4), dpi = 136)
ax1.plot(phi/pi, c, ".-", label = r"$\cos(\phi)$")
ax1.plot(phi/pi, s, ".-", label = r"$\sin(\phi)$")
ax1.plot(phi/pi, s_p, ".-", label = r"$\sin(\phi+\pi/2)$")
ax1.legend() 
ax1.set_xlabel(r"$\phi$ in units of $\pi$")
ax1.set_ylabel(r"amplitude")
ax1.grid(True)
plt.plot()
```

## Euler and Cartesian Representation

$$
z = r e^{i\phi} = |z|e^{i\phi} = r\left(\cos\phi +i\sin\phi \right) = a + ib
$$

$$
|z|^2 = \bar{z}z = (a+ib)(a-ib) = a^2 - (ib)^2 = a^2 - i^2b^2 = a^2 -(-1)b^2 = a^2 + b^2
$$

```python
z = c + s*1j
```

```python
re = real(z)
im = imag(z)
#print(f"Re: {re}")
#print(f"Im: {im}")
#print(z)
```
