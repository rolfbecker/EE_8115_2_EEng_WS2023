```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import log10, log
```

```python
x = np.arange(1,100.001,0.1)
```

```python
y1 =   10*x**2
y2 =  100*x**2
y3 = 1000*x**2
```

```python
fig1,ax1 = plt.subplots()
ax1.plot(x,y1,label=r"$y_1 = 10x^2$")
ax1.plot(x,y2,label=r"$y_2 = 100x^2$")
ax1.plot(x,y3,label=r"$y_3 = 1000x^2$")
ax1.set_title("Linear-Linear Plots (linear axes scaling)")
ax1.legend()
ax1.grid(True)
ax1.set_xlabel(r"$x$")
ax1.set_ylabel(r"$y$")
plt.plot()
```

```python
fig2,ax2 = plt.subplots()
ax2.loglog(x,y1,label=r"$y_1 = 10x^2$")
ax2.loglog(x,y2,label=r"$y_2 = 100x^2$")
ax2.loglog(x,y3,label=r"$y_3 = 1000x^2$")
ax2.set_title("Log-Log Plots (log axes scaling, decadic logarithm)")
ax2.legend()
ax2.grid(True)
ax2.set_xlabel(r"$x$")
ax2.set_ylabel(r"$y$")
plt.plot()
```

```python
fig3,ax3 = plt.subplots()
ax3.plot(log10(x),log10(y1),label=r"$y_1 = 10x^2$")
ax3.plot(log10(x),log10(y2),label=r"$y_1 = 100x^2$")
ax3.plot(log10(x),log10(y3),label=r"$y_1 = 1000x^2$")
ax3.set_title("Linear-Linear Plots (linear axes scaling of $\log_{10}(\ldots)$ values)")
ax3.legend()
ax3.grid(True)
ax3.set_xlabel(r"$\log_{10}(x)$")
ax3.set_ylabel(r"$\log_{10}(y)$")
plt.plot()
```

$$
\begin{eqnarray}
& & \log(y)  =  m\log(x) + b \\
& \Leftrightarrow & y = 10^{m\log(x) + b} \\
& \Leftrightarrow & y = 10^{m\log(x)}\cdot 10^b \\
& \Leftrightarrow & y = \left(10^{\log(x)}\right)^m \cdot 10^b \\
& \Leftrightarrow & y = 10^b x^m \\
\end{eqnarray}
$$
